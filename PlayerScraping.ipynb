```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.worldfootball.net/players_list/eng-premier-league-2019-2020/nach-name/1/"

# Make a GET request to fetch the raw HTML content
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content, "lxml")
```

```python
soup1 = soup.findAll('table')[1].findAll('tr')[0].findAll('th')
print(soup1)
```

```python
import numpy as np
dunkelSoup = soup.findAll('td',class_='dunkel')
hellSoup = soup.findAll('td',class_='hell')
myArr = np.array([])
for wrapper in dunkelSoup:
    print(str(wrapper.text).replace('\n',''))
    print('---')
    
```

```python
for wrapper in hellSoup:
    print(wrapper.text)
```
